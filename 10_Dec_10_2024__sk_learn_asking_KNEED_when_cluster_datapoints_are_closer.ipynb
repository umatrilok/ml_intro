```python
import pandas as pd
import numpy as np
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from kneed import KneeLocator
```

# make_blobs is going to generate random numbers --------> everytime we execute make_blobs it gives different clusters----> how many ever clusters we ask for 

```python
data = make_blobs(n_samples = 200, n_features=2,centers =2)[0]
data
```

```python
#lets convert into dataframe
```

```python
df = pd.DataFrame(data,columns=['age','height'])
df.head()
```

```python
plt.scatter(x='age',y='height',data=df)
plt.show()
```

```python
inertia = []
for k in range(1,11):
    model = KMeans(n_clusters=k)
    model.fit(df)
    inertia.append(model.inertia_)
plt.plot(range(1,11),inertia,marker='o')
plt.show()
```

## Perfect elbow point is comming from the graph

```python
kl = KneeLocator(x=range(1,11),y=inertia,curve='convex',direction='decreasing')
kl.elbow
```

# Kneed is also saying 2 optimal clusters 

```python
# We can make 3 clusters also ----> Make CENTERS = 3 
```

```python
#-------> when 4 clusters are taken separation is not that great b/w the clusters, so we are taking 3
```

# When we take data for 5 Clusters

```python
data = make_blobs(n_samples = 200, n_features=2,centers =5)[0]
data
```

```python
df = pd.DataFrame(data,columns=['age','height'])
df.head()
```

```python
plt.scatter(x='age',y='height',data=df)
plt.show()
```

```python
inertia = []
for k in range(1,11):
    model = KMeans(n_clusters=k)
    model.fit(df)
    inertia.append(model.inertia_)
plt.plot(range(1,11),inertia,marker='o')
plt.show()
```

# point 34 -->Perfect elbow point is "not coming" from the graph  --> it can be 3 or 4 

```python
# Perfect elbow point is not coming because clusters are little closer
```

```python
model = KMeans(n_clusters = 5)
model.fit(df)
clusters = model.predict(df)
centers = model.cluster_centers_
print(model.inertia_)
```

```python
plt.scatter(x='age',y='height',data=df[clusters==0],label='Cluster 0')
plt.scatter(x='age',y='height',data=df[clusters==1],label='Cluster 1')
plt.scatter(x='age',y='height',data=df[clusters==2],label='Cluster 2')
plt.scatter(x='age',y='height',data=df[clusters==3],label='Cluster 3')
plt.scatter(x='age',y='height',data=df[clusters==4],label='Cluster 4')
plt.scatter(centers[:,0],centers[:,1],c='red',s=100)
```

# Lets Ask the KNEED locator

# 4 Optimal clusters --> we 

```python
kl = KneeLocator(x=range(1,11),y=inertia,curve='convex',direction='decreasing')
kl.elbow
```

# Like explained in point 34 it should be 3 or 4 clusters from the graph not 5 for this set of data points --> from graph also we can see that orange and purple clusters are together--> so they can become one cluster
# and from the KNEED locator 4 clusters is optimal
# dont train using 5 clusters but train using 4 clusters

# now we know that 4 clusters are better --> we are traning with 4 clusters

```python
model = KMeans(n_clusters = 4)
model.fit(df)
clusters = model.predict(df)
centers = model.cluster_centers_
print(model.inertia_)
```

```python
plt.scatter(x='age',y='height',data=df[clusters==0],label='Cluster 0')
plt.scatter(x='age',y='height',data=df[clusters==1],label='Cluster 1')
plt.scatter(x='age',y='height',data=df[clusters==2],label='Cluster 2')
plt.scatter(x='age',y='height',data=df[clusters==3],label='Cluster 3')
plt.scatter(centers[:,0],centers[:,1],c='red',s=100)
```

```python
# We Started with 5 clusters, but 4 clusters are better for this set of data
```
