```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier #only this is needed here 

from sklearn.tree import plot_tree
#from sklearn.neighbors import KNeighborsClassifier
#from sklearn.linear_model import LinearRegression,LogisticRegression,RidgeClassifier
```

```python
fruits = pd.read_csv('fruits.csv')
fruits.shape
```

```python
fruits.head()
```

```python
plt.scatter(x='sphericity',y='weight',data=fruits[fruits.label=='Apple'],c='red',label='Apple')
plt.scatter(x='sphericity',y='weight',data=fruits[fruits.label=='Orange'],c='green',label='Orange')
```

```python
model = DecisionTreeClassifier(criterion = 'entropy')#Improt the class DecisionTreeClassifier, create the instance(model) of the class
type(model)
```

```python
# Model is nothing but an instance of a class for implementing perticular machine learning algorithm
```

```python
X = fruits[['sphericity','weight']]
y = fruits['label']
```

```python
model.fit(X,y)
```

```python
#We want to visualize the tree
```

```python
plot_tree(model,feature_names=['sphericity','weight'],class_names=['Apple','Orange'])
plt.show()
#It is alphabetical #Apples and Oranges
```

```python
plt.figure(figsize=[6,8])
plot_tree(model,feature_names=['sphericity','weight'],class_names=['Apple','Orange'])#x[0] = 'sphericity' x[1]='weight'
plt.show()
```

```python
plt.figure(figsize=[6,8])
plot_tree(model)#x[0] = 'sphericity' x[1]='weight'
plt.show()
```

```python
plt.figure(figsize=[6,8])
plot_tree(model,feature_names=['sphericity','weight'],class_names=['Apple','Orange'],filled=True,rounded=True,node_ids=True)#filled = True for adding colors
plt.show()
```

```python
#Let us draw a boundary at 138
#get the scatter plot
plt.scatter(x='sphericity',y='weight',data=fruits[fruits.label=='Apple'],c='red',label='Apple')
plt.scatter(x='sphericity',y='weight',data=fruits[fruits.label=='Orange'],c='green',label='Orange')
plt.plot([0.7,0.95],[138,138],c='blue') #drawing the boundary at 138
plt.plot([0.7,0.95],[150.5,150.5],c='blue') #drawing the boundary at 138
plt.plot([0.796,0.796],[138,150.5],c='green')#drawing the vertical line
```
