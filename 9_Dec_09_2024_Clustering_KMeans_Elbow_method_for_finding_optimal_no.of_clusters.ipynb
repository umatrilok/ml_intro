```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import minmax_scale
```

```python
X = pd.read_csv("fruits.csv",usecols=['sphericity','weight'])
X_scaled = pd.DataFrame(minmax_scale(X),columns = X.columns)
X_scaled
```

```python
##### How to findout the optimal number of clusters ? Ans: Calculate INERTIA
```

```python
####### In the last class we found out INERTIA manually ----> we calculated the centers manually 
```

```python
############################ THIS IS NOT MANUAL CALULATION FOR INERTIA ####################################
```

```python
############ this is for k=1 to 11--------> 1 cluster to 11 clusters

inertia = []
for k in range(1,11):
    model = KMeans(n_clusters=k)
    model.fit(X)
    inertia.append(model.inertia_)
inertia
```

```python
X_scaled
```

```python
#This is very small data, scaled data we cannot identify the knee/elbow point
#---> look at the data it is very small and there is very small decrease in INERTIA ----->so better go with not scaled data
```

```python
############ this is for k=1 to 11--------> 1 cluster to 11 clusters

inertia_scaled = []
for k in range(1,11):
    model = KMeans(n_clusters=k)
    model.fit(X_scaled)
    inertia_scaled.append(model.inertia_)
inertia_scaled
```

```python
# When k increases inertia decreases
```

```python
# Where INERTIA decreases slowly[Instead of Drastical Decrease] --> that is the optimal value of k(no.of clusters)
```

```python
from kneed import KneeLocator
```

```python
kl = KneeLocator(x=range(1,11),y=inertia,curve='convex',direction='decreasing')
kl.elbow
```

```python
# Generally we get the confusion if the optimal cluster's are 2 or if the optimal cluster's are 3,
# --> confusion is b/w 2 or 3 consicutive numbers. ---> it wont be like 3 or 30 --> it will b like 3 or 4

# Identifying the Optimal no.of clusters is going to be the main concern.
```

```python
# From the ELBOW point it is clear that 3 clusters are better for unscaled data 
model = KMeans(n_clusters = 3)
model.fit(X)
clusters = model.predict(X)
centers = model.cluster_centers_
```

```python
plt.scatter(x='sphericity',y='weight',data=X[clusters==0],c='red', label='Cluster 0')
plt.scatter(x='sphericity',y='weight',data=X[clusters==1],c='green', label='Cluster 1')
plt.scatter(x='sphericity',y='weight',data=X[clusters==2],c='orange', label='Cluster 2')
# all row's 0th column ==> [:,0]
# all row's 1st column ==> [:,1]
plt.scatter(x=centers[:,0],y=centers[:,1],c='blue',s=100)
plt.legend()
plt.show()
```

We are not saying all the red color one's are apples
We are not saying all the green color one's are oranges

We are SAYING all the red color one's have Similar Characters 
We are SAYING all the blue color one's have Similar Characters
