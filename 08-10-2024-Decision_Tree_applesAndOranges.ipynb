```python
# Can we separate apples and oranges with only sphericity/only weight ?
# Yes we can 90%. Weight which are less than 140 are Oranges. If we clearly observer there are some more Oranges which are over 140 
```

```python
# When we combine Weight and Sphericity, we can separate with higher accuracy
```

```python
plt.scatter(x='sphericity',y='weight',data=fruits[fruits.label=='Apple'],c='red',label='Apple')
```

```python
plt.scatter(x='sphericity',y='weight',data=fruits[fruits.label=='Orange'],c='green',label='Orange')
```

```python
#combining above two statements
plt.scatter(x='sphericity',y='weight',data=fruits[fruits.label=='Apple'],c='red',label='Apple')
plt.scatter(x='sphericity',y='weight',data=fruits[fruits.label=='Orange'],c='green',label='Orange')
plt.legend()
plt.xlabel('sphericty')
plt.ylabel('weight')
plt.show()
```
